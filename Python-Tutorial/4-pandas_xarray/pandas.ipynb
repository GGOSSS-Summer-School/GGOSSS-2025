```python
def std_perso(data):
    return np.sqrt(((data-data.mean())**2).mean())

print(penguins.groupby('sex').bill_length_mm.apply(std_perso))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

correlation = penguins[['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']].corr()

fig, ax = plt.subplots(figsize=(12,8))
sns.heatmap(correlation, 
            cmap="seismic",#viridis    
            square=True, 
            linewidth=2, 
            annot=True,
            fmt = '0.2f',
           )
plt.show()
```

```python

b2 = boats[['engine_age','length','power','month','weight','value','value_cpi','price','year','dist','y_month','patch_area','nao_index']]

"""
correlation = b2.loc[b2.power > 400]\
                .groupby('ID')\
                .mean()\
                .corr()
"""

correlation = b2.corr()
correlation = np.round(correlation,2)
#np.ceil()
#np.floor()

# Mask only upper triangle (excluding/including diagonal: k=1/k=0)
mask = np.triu(np.ones_like(correlation, dtype=bool),k=1)

# Mask only the diagonal
#mask = np.eye(len(correlation), dtype=bool)

fig, ax = plt.subplots(figsize=(12,8), dpi=100)
sns.heatmap(correlation, 
            mask=mask,
            cmap="bwr_r", #"seismic", #viridis    
            square=True, 
            linewidth=2, 
            annot=True,
            center=0,
            vmin = -1,
            vmax = 1,
            #fmt = '0.2f',
            cbar_kws={'label': 'Correlation','orientation':'vertical','shrink':.6, 'pad':.03},
            #cbar_kws={'label': 'Correlation', 'shrink': 0.8, 'ticks': [-1,-.75,-.5,-0.25,0,.25,.5,.75,1]}
)

plt.title("Norvegian fishing boat's features")

format_save = ['.png','.eps','.jpeg','.jpg','.pdf']
for extension in format_save:
    plt.savefig('norvegian_boats_features_coorelation'+extension ,dpi =fig.dpi,bbox_inches='tight',transparent=False)

plt.show()
```

```python
weather = pd.read_csv('data/weather_dataset.csv')
weather
```

```python
weather['Date'] = pd.to_datetime(weather['Date/Time'])
weather = weather.set_index(weather.Date)
weather
```

```python
weather.Temp_C.plot()
```

```python
weather.Temp_C.loc['05-2012'].plot() ## dd-mm-yyyy   yyyy-mm-dd
```

```python
weather.Temp_C.loc['01-01-2012':'15-03-2012'].plot()
plt.title('Temperature')
```

```python
fig = plt.figure(figsize=(10,6))
weather.Temp_C.plot(label='donnee brute',linewidth=1)
weather.Temp_C.resample('ME').mean().plot(label='moyenne mois')
weather.Temp_C.resample('D').mean().plot(label='moyenne jour')
plt.legend(ncol=3)
```

```python
weather.Temp_C.rolling(12,center=True).mean().plot()
weather.Temp_C.rolling(12,center=True).std().plot()
```
