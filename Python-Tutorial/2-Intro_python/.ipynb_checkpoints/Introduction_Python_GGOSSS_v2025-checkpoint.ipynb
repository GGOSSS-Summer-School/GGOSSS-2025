Dans un premier temps, on se contentera d'utiliser des "matrices" (**array** de numpy de dimension > 1) et des "vecteurs" (array de dimension 1) pour coder en Python l'**algèbre** que vous connaissez...

```python
import numpy as np
print(np.eye(4)) # Identité de rang p=4 ici
x=np.arange(3)
M=np.arange(9)
M=np.reshape(M,(3,3))
print(M)
print(M.shape)
print(np.dot(M,x))
print(np.dot(np.transpose(x),x))
print(np.outer(x, np.transpose(x))) # attention avec 'dot'...
```

Numpy (ainsi que d'autres modules) sont organisés en **sous-ensembles** (ci-après **linalg** ; voir https://numpy.org/doc/stable/reference/routines.linalg.html) rassemblant des fonctionnalités spécifiques (ici, de nombreux outils d'**algèbre linéaire** courante).

```python
import numpy as np
from numpy.linalg import inv  # pour disposer de l'inversion de matrice de linalg
M=np.random.randn(3,3)

print(M)
print(" Inverser une matrice... et vérifier le produit avec la matrice d'origine")
print(np.dot(inv(M),M))
```

**ENUMERATE & ZIP**

fonction utilisées avec les boucles

```python
#%%%
number = np.arange(10)
alphabet = ['a','b','c','d','e','f','g','h','i','j']
print(alphabet)

#1
for i,alpha in enumerate(alphabet):
    print(f'index is: {i}, letter is {alpha}')

#2  
for _,alpha in enumerate(alphabet):
    print(f'index is: , letter is {alpha}')

#3
for num,alpha in zip(number,alphabet):
    print(f'number is: {num}, letter is {alpha}')
#4
for i,(num,alpha) in enumerate(zip(number, alphabet)):
    print(f'index is {i}, number is: {num}, letter is {alpha}')
```

```python
import numpy as np

R = np.random.randint(1, 42, 100)  # tire 100 entiers dans [1, 42]; see numpy.random
print("Initial array: ",R)

PaiR = []   # crée une liste vide

for e in R:
    if (e%2 == 0):
        PaiR.append( e )  # y stocke les valeurs paires de R
        
print("List of even elements: ",PaiR)
print("Sorted... ", sorted(PaiR) )

X=[4, 3, 2, 1]
X.sort()  # méthode sur liste
print(X)

for e in X:
    print(e)
```

**(6) Fonctions** (et **tuple**...)

Une "fonction" peut i) avoir ou pas d'arguments en entrée et ii) retourner ou pas de valeurs...

```python
def echange(a,b):
    c=a
    a=b
    b=c
    print(a,b)
    
# fin de fonction

echange(9,2)
```

Une fonction avec une commande **return** pour récupérer les éléments voulus (éventuellement multiples, voire plus complexes que des nombres) après appel

```python
def echange(a,b):
    c=a
    a=b
    b=c
    return a,b
# fin de fonction

print(echange(9,2))

# ou encore...
x, y = echange(9, 2)
print(x,y)

# maintenant utilisation d'un "tuple" (objet itérable, aussi...)
TPL = echange(1, 42)
print(TPL, TPL[0], TPL[1])
```

**btw** une autre possibilité offerte : affectations *simultanées*...

```python
a = 1
b = 42
print(a, b)
a, b = b, a
print(a, b)
```

```python
def Determinant(A):
    from numpy.linalg import det
    return det(A)

import numpy as np
A=np.identity(3)
print('Déterminant de :')
print(A,' = ',Determinant(A))
```

Un graphique élémentaire à partir d'un fichier **ascii** : https://gml.noaa.gov/webdata/ccgg/trends/co2/co2_annmean_mlo.txt (si vous ne voulez *pas* l'enregistrer, passez à la cellule suivante...)

```python
import numpy as np
import matplotlib.pyplot as plt

#plt.xkcd()
#--- ouvre le fichier ascii
f=open('../co2_annmean_mlo.txt','r')  # sinon passez à la cellule suivante...

#--- tout lire
tout=f.readlines()
nlines=len(tout)

#--- fermer le fichier
f.close()

#--- initialiser deux listes : year/taux co2

year=[]
tco2=[]

#--- eliminer les 61 premieres lignes de commentaire
for i in np.arange(62,nlines):
    year.append( float( str.split(tout[i])[0] ) )
    tco2.append( float(str.split(tout[i])[1]) )

plt.plot(year, tco2, 'k')
plt.title('Mauna Loa (HI) data')
plt.xlabel('year', fontsize=16)
plt.ylabel('CO2 concentration (ppm)', fontsize=16)
plt.show()
```

Un dernier exemple avec l'utilisation de la **FFT** sur un signal *x(t)*...

```python
import matplotlib.pyplot as plt
import numpy as np

# fourier material now...
from numpy.fft import fft, ifft

#import warnings
#warnings.filterwarnings("ignore")

# sampling rate
sr = 2000
# sampling interval
ts = 1.0/sr
t = np.arange(0,1,ts)

freq = 3.
x = 3*np.sin(2*np.pi*freq*t) + np.sin(2*np.pi*(freq+2)*t)

X = fft(x)
N = len(X)
n = np.arange(N)
T = N/sr
freq = n/T 

plt.figure(figsize = (6, 3))
plt.subplot(121)
plt.plot(t, ifft(X), 'r')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.subplot(122)
plt.stem(freq, np.abs(X), 'k', markerfmt="o", basefmt="-k")  # voir doc stem
plt.xlabel('Freq (Hz)')
plt.ylabel('FFT Amplitude |X(freq)|')
plt.xlim(0,10)

plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0)  # voir doc tight_layout

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-4, 4, 91)
y = np.linspace(-4, 4, 91)

Im = np.zeros((x.size, y.size))
for i in np.arange(x.size):
    for j in np.arange(y.size):
        Im[i, j] = np.sinc(x[i]) * np.sinc(y[j])

plt.hot()
plt.imshow(Im)
plt.colorbar()
plt.show()
```

```python
plt.contour(x, y, Im)
plt.show()
```

```python
plt.contourf(x, y, Im)
plt.show()
```
